```python
data=data.iloc[600000:1100000,:]
```

```python
import nltk
nltk.download('stopwords')
port_stem = PorterStemmer()
```

```python
def stemming(text):
    text = re.sub('[^a-zA-Z]',' ',text)
    text = text.lower()
    text = text.split()
    text = [port_stem.stem(word) for word in text if not word in stopwords.words('english')]
    text = ' '.join(text)
    return text
```

```python
data
```

```python
data['text'] = data['text'].apply(stemming)
```

```python
data['text']
```

```python
#splitting X AND Y
X = data['text']
Y = data['target']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
vectorizer=TfidfVectorizer()

X_train=vectorizer.fit_transform(X_train)
X_test=vectorizer.transform(X_test)
```

```python
model=LogisticRegression(max_iter=1000)
```

```python
model.fit(X_train,Y_train)
```

```python
X_train_pred=model.predict(X_train)
train_data_acc=accuracy_score(X_train_pred,Y_train)
```

```python
X_test_pred=model.predict(X_test)
train_data_acc=accuracy_score(X_test_pred,Y_test)
```
